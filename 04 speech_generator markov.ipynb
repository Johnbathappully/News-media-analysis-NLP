```python
def generateTable (text):
    t={}
    text=text.split()
    for i in range(len(text)-1):
        x=text[i]
        y=text[i+1]
        
        if t.get(x) is None:
            t[x] ={}
            t[x][y]=1
        else:
            if t[x].get(y) is None:
                t[x][y]=1
            else:
                t[x][y] +=1
    return t
```

```python
def convertFreqIntoProb(t):
    for kx in t.keys():
        s=float(sum(t[kx].values()))
        for k in t[kx].keys():
            t[kx][k] =t[kx][k]/s
    return t
```
